### Multinomial logistic regression model

The model applies the softmax function (Eq. 3.13) after a linear transformation to produce class probabilities:

$$ \hat{\mathbf{y}}^{(i)} = \text{softmax}(\mathbf{W}\mathbf{x}^{(i)} + \mathbf{b}) $$

The cross-entropy cost function (Eq. 3.16) over $N$ training samples with $K$ classes is:

$$ J = -\frac{1}{N} \sum_{i=1}^{N} \sum_{k=1}^{K} y_k^{(i)} \log \hat{y}_k^{(i)} $$

Since the gradient of the softmax–cross-entropy combination simplifies to $\hat{\mathbf{y}} - \mathbf{y}$ (Eq. 3.32), the parameter gradients are:

$$ \frac{\partial J}{\partial \mathbf{W}} = \frac{1}{N} \sum_{i=1}^{N} (\hat{\mathbf{y}}^{(i)} - \mathbf{y}^{(i)}) \cdot (\mathbf{x}^{(i)})^T, \qquad \frac{\partial J}{\partial \mathbf{b}} = \frac{1}{N} \sum_{i=1}^{N} (\hat{\mathbf{y}}^{(i)} - \mathbf{y}^{(i)}) $$

Note: This notebook implements batch gradient descent over the full training set, rather than the per-sample SGD formulation.
